```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('train.csv')
count_na_b = df.isna().sum()
```

```python
cols_to_plot = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']

fig, axes = plt.subplots(2, 3, figsize=(12, 8))
axes = axes.flatten()

for i, col in enumerate(cols_to_plot):
    sns.histplot(data=df, x=col, bins=30, ax=axes[i])

    mean = df[col].mean()
    axes[i].axvline(mean, color='red', label='Mean')
    axes[i].legend()
    axes[i].set_xlabel('')

plt.tight_layout()
plt.show()
```

```python
df.info()
```

```python
mean_val_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_val_age)

df.info()
```

```python
df['Age_group'] = np.select([df['Age'] <= 21, df['Age'] <= 44, df['Age'] <= 59, df['Age'] >= 60],
                      ["teenager", "young adult", "middle-aged", "elderly"],
                      default=None)
```

```python
df.head(20)
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.info()
```

```python
df['Cabin_class'] = df['Cabin_class'] = df['Cabin'].str[0]
df['Cabin_class'] = df['Cabin_class'].fillna("None")


```

```python
df.head(10)
```

```python
df['Pclass'] = np.select([df['Pclass'] == 1, df['Pclass'] == 2, df['Pclass'] == 3],
                         ['F', 'S', 'T'],
                         default=None)
```

```python
for i, Name in enumerate(df['Name']):
  Name = Name.split(' ')
  for j in Name:
    if j.endswith('.'):
      rank = j
  df.loc[i, 'title'] = rank
```

```python
df['Sex'] = np.where(df['Sex'] == 'male', 1, 0)
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df['IsAlone'] = np.where(df['FamilySize'] == 1, 1, 0)
```

```python
sns.boxplot(df['Fare'])
```

```python
IQR = df['Fare'].quantile(0.75) - df['Fare'].quantile(0.25)
lowest = df['Fare'].quantile(0.25) - IQR * 1.5
highest = df['Fare'].quantile(0.75) + IQR * 1.5

df['Fare'] = df['Fare'].clip(lowest, highest)
```

```python
sns.boxplot(df['Fare'])
```

```python
sns.histplot(df['Age'])
```

```python
df['Age'] = df['Age'].clip(upper=df['Age'].quantile(0.95))
```

```python
sns.histplot(df['Age'])
```
